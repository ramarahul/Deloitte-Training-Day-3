```python
teams=['Team1','Team2','Team3','Team4','Team5','Team6','Team7','Team8','Team9','Team10','Team11','Team12','Team13','Team14','Team15','Team16']
a=['Team1','Team2','Team3','Team4']
b=['Team5','Team6','Team7','Team8']
c=['Team9','Team10','Team11','Team12']
d=['Team13','Team14','Team15','Team16']


def whoWon(a):
    '''returns a list of winners from a list of 4'''
    won={}
    winners=[]
    #Arranging matches and populating won dictionary
    for i in range(3):
        for j in range(i+1,4):
            print('who won between',a[i],'and',a[j])
            winner=input()
            if winner=='draw':
                won[a[i]]=won[a[i]]+1
                won[a[j]]=won[a[j]]+1
            elif winner==a[i]:
                if a[i] in won:
                    won[a[i]]=won[a[i]]+2
                else:
                    won[a[i]]=2
            elif winner==a[j]:
                if a[j] in won:
                    won[a[j]]=won[a[j]]+2
                else:
                    won[a[j]]=2
    count=0
    for k in sorted(won, key=won.get, reverse=True):     #Sorting the dictionary to get winners
           if count<2: 
                winners.append(k)
                count=count+1
    return winners
#First Round of Tournament
wina=whoWon(a)
winb=whoWon(b)
winc=whoWon(c)
wind=whoWon(d)
#Second Round of Tournament 
e=wina+winb
f=winc+wind
wine=whoWon(e)
winf=whoWon(f)
#Final Round of Tournament
g=wine+winf
final=whoWon(g)
#Print the final winner
print("Final winner is",final[0])
        
```
